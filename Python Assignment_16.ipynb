```python
# 1. Create a list called years_list, starting with the year of your birth, and each year thereafter until your fifth birthday.
birth_year = 1990  # Example year, change it to your actual birth year
years_list = [birth_year + i for i in range(6)]  # 6 years including birth year and the fifth birthday year
print(years_list)

# 2. In which year in years_list was your third birthday?
third_birthday_year = years_list[3]  # 0-based indexing, so index 3 corresponds to the third birthday
print(third_birthday_year)

# 3. In the years list, which year were you the oldest?
oldest_year = years_list[-1]  # The last year in the list is the oldest
print(oldest_year)

# 4. Make a list called things with these three strings as elements: "mozzarella", "cinderella", "salmonella".
things = ["mozzarella", "cinderella", "salmonella"]
print(things)

# 5. Capitalize the element in things that refers to a person and then print the list. Did it change the element in the list?
things[1] = things[1].capitalize()  # Capitalizing "cinderella"
print(things)

# 6. Make a surprise list with the elements "Groucho," "Chico," and "Harpo."
surprise = ["Groucho", "Chico", "Harpo"]
print(surprise)

# 7. Lowercase the last element of the surprise list, reverse it, and then capitalize it.
last_element = surprise[-1].lower()[::-1].capitalize()
print(last_element)

# 8. Make an English-to-French dictionary called e2f and print it.
e2f = {"dog": "chien", "cat": "chat", "walrus": "morse"}
print(e2f)

# 9. Write the French word for walrus in your three-word dictionary e2f.
walrus_french = e2f["walrus"]
print(walrus_french)

# 10. Make a French-to-English dictionary called f2e from e2f. Use the items method.
f2e = {value: key for key, value in e2f.items()}
print(f2e)

# 11. Print the English version of the French word chien using f2e.
english_for_chien = f2e["chien"]
print(english_for_chien)

# 12. Make and print a set of English words from the keys in e2f.
english_words_set = set(e2f.keys())
print(english_words_set)

# 13. Make a multilevel dictionary called life.
life = {
    'animals': {
        'cats': ['Henri', 'Grumpy', 'Lucy'],
        'octopi': [],
        'emus': []
    },
    'plants': {},
    'other': {}
}
print(life)

# 14. Print the top-level keys of life.
print(life.keys())

# 15. Print the keys for life['animals'].
print(life['animals'].keys())

# 16. Print the values for life['animals']['cats']
print(life['animals']['cats'])
```
