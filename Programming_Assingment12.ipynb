```python
# 1. Extract Unique values from dictionary values
def extract_unique_values(dictionary):
    unique_values = set()
    for value in dictionary.values():
        if isinstance(value, list):
            unique_values.update(value)
        else:
            unique_values.add(value)
    return unique_values

# 2. Find the sum of all items in a dictionary
def sum_of_items(dictionary):
    return sum(dictionary.values())

# 3. Merging two Dictionaries
def merge_dictionaries(dict1, dict2):
    merged = dict1.copy()  # Make a copy of the first dictionary
    merged.update(dict2)  # Update with the second dictionary
    return merged

# 4. Convert key-values list to flat dictionary
def list_to_dict(key_value_list):
    return dict(key_value_list)

# 5. Insertion at the beginning in OrderedDict
from collections import OrderedDict

def insert_at_beginning(ordered_dict, key, value):
    # Create a new OrderedDict with the new item inserted at the beginning
    new_ordered_dict = OrderedDict([(key, value)] + list(ordered_dict.items()))
    return new_ordered_dict

# 6. Check order of character in string using OrderedDict
def check_order_of_characters(string, pattern):
    ordered_dict = OrderedDict.fromkeys(string)
    pattern_index = 0
    for char in ordered_dict:
        if char == pattern[pattern_index]:
            pattern_index += 1
        if pattern_index == len(pattern):
            return True
    return False

# 7. Sort Python Dictionaries by Key or Value
def sort_dict(dictionary, by_value=False):
    if by_value:
        return {k: v for k, v in sorted(dictionary.items(), key=lambda item: item[1])}
    else:
        return {k: v for k, v in sorted(dictionary.items())}

# Example usage:

# 1. Extract Unique values from dictionary values
example_dict = {'a': [1, 2, 3], 'b': [3, 4, 5], 'c': 6}
print("Unique values:", extract_unique_values(example_dict))

# 2. Find the sum of all items in a dictionary
sum_dict = {'a': 10, 'b': 20, 'c': 30}
print("Sum of all items:", sum_of_items(sum_dict))

# 3. Merging two Dictionaries
dict1 = {'a': 1, 'b': 2}
dict2 = {'c': 3, 'd': 4}
print("Merged Dictionary:", merge_dictionaries(dict1, dict2))

# 4. Convert key-values list to flat dictionary
key_value_list = [('a', 1), ('b', 2), ('c', 3)]
print("Dictionary from key-value list:", list_to_dict(key_value_list))

# 5. Insertion at the beginning in OrderedDict
ordered_dict = OrderedDict([('a', 1), ('b', 2)])
new_ordered_dict = insert_at_beginning(ordered_dict, 'z', 0)
print("OrderedDict with insertion at the beginning:", new_ordered_dict)

# 6. Check order of character in string using OrderedDict
string = "abcde"
pattern = "ace"
print("Is the pattern present in order:", check_order_of_characters(string, pattern))

# 7. Sort Python Dictionaries by Key or Value
dict_to_sort = {'a': 3, 'c': 1, 'b': 2}
print("Sorted by Key:", sort_dict(dict_to_sort, by_value=False))
print("Sorted by Value:", sort_dict(dict_to_sort, by_value=True))
```
