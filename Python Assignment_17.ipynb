```python
# 1. Assign the value 7 to the variable guess_me. Write conditional tests.
guess_me = 7
if guess_me < 7:
    print('too low')
elif guess_me > 7:
    print('too high')
else:
    print('just right')

# 2. Assign the value 7 to guess_me and 1 to start. Write a while loop.
guess_me = 7
start = 1
while start != guess_me:
    if start < guess_me:
        print('too low')
        start += 1
    elif start > guess_me:
        print('oops')
        break
else:
    print('found it!')

# 3. Print the following values of the list [3, 2, 1, 0] using a for loop.
for number in [3, 2, 1, 0]:
    print(number)

# 4. Use a list comprehension to make a list of the even numbers in range(10).
even_numbers = [x for x in range(10) if x % 2 == 0]
print(even_numbers)

# 5. Use a dictionary comprehension to create the dictionary squares.
squares = {x: x**2 for x in range(10)}
print(squares)

# 6. Construct the set odd from the odd numbers in the range using a set comprehension.
odd_numbers = {x for x in range(10) if x % 2 != 0}
print(odd_numbers)

# 7. Use a generator comprehension to return the string 'Got ' and a number for the numbers in range(10).
got_numbers = ('Got ' + str(x) for x in range(10))
for item in got_numbers:
    print(item)

# 8. Define a function called good that returns the list ['Harry', 'Ron', 'Hermione'].
def good():
    return ['Harry', 'Ron', 'Hermione']
print(good())

# 9. Define a generator function called get_odds that returns the odd numbers from range(10).
def get_odds():
    for x in range(10):
        if x % 2 != 0:
            yield x

# Use a for loop to find and print the third value returned by get_odds.
odd_gen = get_odds()
third_value = None
for i, value in enumerate(odd_gen):
    if i == 2:  # third value (index 2)
        third_value = value
        break
print(third_value)

# 10. Define an exception called OopsException. Raise this exception to see what happens.
class OopsException(Exception):
    pass

try:
    raise OopsException('Oops! Something went wrong.')
except OopsException as e:
    print(f'Caught an oops: {e}')

# 11. Use zip() to make a dictionary called movies.
titles = ['Creature of Habit', 'Crewel Fate']
plots = ['A nun turns into a monster', 'A haunted yarn shop']
movies = dict(zip(titles, plots))
print(movies)
```
