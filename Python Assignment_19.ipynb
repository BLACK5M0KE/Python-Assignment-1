```python
# Task 1: Create a class called Thing with no contents and print it.
class Thing:
    pass

# Print the class itself
print(Thing)

# Create an object called example from the Thing class and print it.
example = Thing()
print(example)

# Task 2: Create a new class called Thing2 and add the value 'abc' to the letters class attribute. Print it.
class Thing2:
    letters = 'abc'

# Print the class attribute 'letters'
print(Thing2.letters)

# Task 3: Make a class Thing3, and assign 'xyz' to an instance attribute called letters. Print it.
class Thing3:
    def __init__(self):
        self.letters = 'xyz'

# Create an object and print its instance attribute 'letters'
thing3 = Thing3()
print(thing3.letters)

# Task 4: Create an Element class with instance attributes name, symbol, and number. Create a class object.
class Element:
    def __init__(self, name, symbol, number):
        self.name = name
        self.symbol = symbol
        self.number = number

# Create an Element object with 'Hydrogen', 'H', and 1
hydrogen = Element('Hydrogen', 'H', 1)
print(hydrogen.name, hydrogen.symbol, hydrogen.number)

# Task 5: Make a dictionary and create an Element object using that dictionary.
element_dict = {'name': 'Hydrogen', 'symbol': 'H', 'number': 1}
hydrogen2 = Element(**element_dict)
print(hydrogen2.name, hydrogen2.symbol, hydrogen2.number)

# Task 6: Define a method dump() to print the object's attributes and call it.
class Element:
    def __init__(self, name, symbol, number):
        self.name = name
        self.symbol = symbol
        self.number = number
    
    def dump(self):
        print(f"Name: {self.name}, Symbol: {self.symbol}, Number: {self.number}")

# Create the hydrogen object and call dump()
hydrogen3 = Element('Hydrogen', 'H', 1)
hydrogen3.dump()

# Task 7: Change method dump to __str__ and print the object.
class Element:
    def __init__(self, name, symbol, number):
        self.name = name
        self.symbol = symbol
        self.number = number
    
    def __str__(self):
        return f"Name: {self.name}, Symbol: {self.symbol}, Number: {self.number}"

# Create a new hydrogen object and print it
hydrogen4 = Element('Hydrogen', 'H', 1)
print(hydrogen4)

# Task 8: Make attributes private and define getter properties.
class Element:
    def __init__(self, name, symbol, number):
        self.__name = name
        self.__symbol = symbol
        self.__number = number
    
    @property
    def name(self):
        return self.__name
    
    @property
    def symbol(self):
        return self.__symbol
    
    @property
    def number(self):
        return self.__number

# Create the hydrogen object and use getters
hydrogen5 = Element('Hydrogen', 'H', 1)
print(hydrogen5.name, hydrogen5.symbol, hydrogen5.number)

# Task 9: Define Bear, Rabbit, and Octothorpe classes with an eats() method.
class Bear:
    def eats(self):
        return 'berries'

class Rabbit:
    def eats(self):
        return 'clover'

class Octothorpe:
    def eats(self):
        return 'campers'

# Create objects and print what they eat.
bear = Bear()
rabbit = Rabbit()
octothorpe = Octothorpe()

print(bear.eats())
print(rabbit.eats())
print(octothorpe.eats())

# Task 10: Define Laser, Claw, and SmartPhone classes with a does() method.
class Laser:
    def does(self):
        return 'disintegrate'

class Claw:
    def does(self):
        return 'crush'

class SmartPhone:
    def does(self):
        return 'ring'

class Robot:
    def __init__(self):
        self.laser = Laser()
        self.claw = Claw()
        self.smartphone = SmartPhone()

    def does(self):
        print(self.laser.does())
        print(self.claw.does())
        print(self.smartphone.does())

# Create a Robot object and print what it does.
robot = Robot()
robot.does()
```
