```python
import logging

# 1. Create an assert statement that throws an AssertionError if the variable spam is a negative integer.
spam = -1
assert spam >= 0, "spam cannot be a negative integer"

# 2. Write an assert statement that triggers an AssertionError if the variables eggs and bacon contain strings that are the same as each other, even if their cases are different.
eggs = "hello"
bacon = "HELLO"
assert eggs.lower() != bacon.lower(), "eggs and bacon contain the same string, ignoring case"

# 3. Create an assert statement that throws an AssertionError every time.
assert False, "This will always raise an AssertionError"

# 4. What are the two lines that must be present in your software in order to call logging.debug()?
import logging

# To enable logging and use logging.debug()
logging.basicConfig(level=logging.DEBUG)  # Set the logging level to DEBUG
logging.debug("This is a debug message")

# 5. What are the two lines that your program must have in order to have logging.debug() send a logging message to a file named programLog.txt?
import logging

# To log messages to a file
logging.basicConfig(filename='programLog.txt', level=logging.DEBUG)  # Log to a file
logging.debug("This message will be logged to programLog.txt")

# 6. What are the five levels of logging?
# The five levels of logging, in order of severity:
# 1. DEBUG
# 2. INFO
# 3. WARNING
# 4. ERROR
# 5. CRITICAL

# 7. What line of code would you add to your software to disable all logging messages?
logging.disable(logging.CRITICAL)  # Disables all logging messages

# 8. Why is using logging messages better than using print() to display the same message?
# Using logging messages is better because:
# - It allows for different levels of severity (DEBUG, INFO, WARNING, etc.).
# - It can output to multiple destinations (console, file, etc.).
# - It can be enabled/disabled based on the logging level.
# - It includes timestamps and other useful information.

# 9. What are the differences between the Step Over, Step In, and Step Out buttons in the debugger?
# - **Step Over**: Executes the current line of code, but if the line contains a function call, it doesn't step into the function. It "skips" over it.
# - **Step In**: Executes the current line of code and steps into any function calls, allowing you to debug inside the function.
# - **Step Out**: Runs the current function to completion and returns to the calling function, stepping out of the current function.

# 10. After you click Continue, when will the debugger stop?
# The debugger will stop when it hits a breakpoint, encounters an exception, or finishes execution of the program.

# 11. What is the concept of a breakpoint?
# A **breakpoint** is a marker set in your code where the debugger will pause the program's execution. This allows you to inspect the state of the program and variables at that specific point in time.
```
