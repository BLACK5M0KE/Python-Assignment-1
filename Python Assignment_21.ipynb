```python
import os
import datetime
import random
import time
import multiprocessing
from datetime import timedelta

# Task 1: Add the current date to the text file today.txt as a string.
today_date = datetime.date.today().strftime('%Y-%m-%d')
with open('today.txt', 'w') as file:
    file.write(today_date)

# Task 2: Read the text file today.txt into the string today_string.
with open('today.txt', 'r') as file:
    today_string = file.read()

# Task 3: Parse the date from today_string.
parsed_date = datetime.datetime.strptime(today_string, '%Y-%m-%d').date()
print(f"Parsed Date: {parsed_date}")

# Task 4: List the files in your current directory.
current_directory_files = os.listdir()
print(f"Files in current directory: {current_directory_files}")

# Task 5: Create a list of all files in your parent directory (minimum five files should be available).
parent_directory_files = os.listdir('..')  # Parent directory
print(f"Files in parent directory: {parent_directory_files[:5]}")  # Limiting to five files

# Task 6: Use multiprocessing to create three separate processes.
def worker():
    time_to_wait = random.randint(1, 5)
    time.sleep(time_to_wait)
    print(f"Current time: {datetime.datetime.now()} - waited for {time_to_wait} seconds")

# Create three processes
processes = []
for _ in range(3):
    p = multiprocessing.Process(target=worker)
    processes.append(p)
    p.start()

# Wait for all processes to complete
for p in processes:
    p.join()

# Task 7: Create a date object of your day of birth.
# Replace 'YYYY-MM-DD' with your actual date of birth.
birth_date = datetime.date(1990, 1, 1)  # Example date: January 1, 1990
print(f"Birth Date: {birth_date}")

# Task 8: What day of the week was your day of birth?
day_of_week = birth_date.strftime('%A')  # Name of the weekday
print(f"Day of the week on birth date: {day_of_week}")

# Task 9: When will you be (or when were you) 10,000 days old?
ten_thousand_days_old = birth_date + timedelta(days=10000)
print(f"10,000 days old will be: {ten_thousand_days_old}")
```
