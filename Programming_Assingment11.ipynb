```python
# 1. Python program to find words which are greater than given length k
def words_greater_than_k(string, k):
    words = string.split()
    return [word for word in words if len(word) > k]

# 2. Python program for removing i-th character from a string
def remove_ith_character(string, i):
    return string[:i] + string[i+1:]

# 3. Python program to split and join a string
def split_and_join(string):
    split_string = string.split()
    return " ".join(split_string)

# 4. Python to check if a given string is binary string or not
def is_binary_string(string):
    return all(char in '01' for char in string)

# 5. Python program to find uncommon words from two Strings
def uncommon_words(str1, str2):
    words1 = set(str1.split())
    words2 = set(str2.split())
    return list((words1 - words2) | (words2 - words1))

# 6. Python to find all duplicate characters in string
def duplicate_characters(string):
    char_count = {}
    duplicates = []
    for char in string:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1
    for char, count in char_count.items():
        if count > 1:
            duplicates.append(char)
    return duplicates

# 7. Python Program to check if a string contains any special character
import re

def contains_special_char(string):
    return bool(re.search(r'[^a-zA-Z0-9\s]', string))

# Example usage

# 1. Words greater than given length
input_string = "The quick brown fox jumps over the lazy dog"
k = 3
print("Words greater than length k:", words_greater_than_k(input_string, k))

# 2. Removing i-th character from a string
string = "Hello, World!"
i = 7
print("String after removing i-th character:", remove_ith_character(string, i))

# 3. Split and join a string
string = "Python is fun"
print("Split and join the string:", split_and_join(string))

# 4. Check if the string is a binary string
binary_string = "1010101"
non_binary_string = "1020101"
print("Is binary string (1010101):", is_binary_string(binary_string))
print("Is binary string (1020101):", is_binary_string(non_binary_string))

# 5. Find uncommon words from two strings
str1 = "Hello world, this is Python."
str2 = "Python is awesome, and Hello world."
print("Uncommon words:", uncommon_words(str1, str2))

# 6. Find duplicate characters in string
string = "programming"
print("Duplicate characters:", duplicate_characters(string))

# 7. Check if a string contains any special character
string = "Hello World!"
print("Contains special character (Hello World!):", contains_special_char(string))
string = "HelloWorld123"
print("Contains special character (HelloWorld123):", contains_special_char(string))
```
