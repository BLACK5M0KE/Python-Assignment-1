```python
# 1. What are escape characters, and how do you use them?
# Escape characters allow special characters to be included in strings using a backslash (\).
example_string = "This is a \"quoted\" word."
print(example_string)  # Output: This is a "quoted" word.

# 2. What do the escape characters \n and \t stand for?
# \n - Newline (moves text to a new line)
# \t - Tab (adds a tab space)
print("Hello\nWorld")  # Newline example
print("Hello\tWorld")  # Tab example

# 3. How to include a backslash character in a string?
backslash_example = "This is a backslash: \\"
print(backslash_example)  # Output: This is a backslash: \

# 4. Why isn't the single quote in "Howl's Moving Castle" a problem?
# The string uses double quotes, so the single quote inside doesn't need escaping.
valid_string = "Howl's Moving Castle"
print(valid_string)  # Output: Howl's Moving Castle

# 5. How to write a string of newlines without using \n?
multiline_string = """Line 1
Line 2
Line 3"""
print(multiline_string)
# Output:
# Line 1
# Line 2
# Line 3

# 6. Values of the given expressions
print('Hello, world!'[1])    # Output: 'e'
print('Hello, world!'[0:5])  # Output: 'Hello'
print('Hello, world!'[:5])   # Output: 'Hello'
print('Hello, world!'[3:])   # Output: 'lo, world!'

# 7. Values of the following expressions
print('Hello'.upper())           # Output: 'HELLO'
print('Hello'.upper().isupper()) # Output: True
print('Hello'.upper().lower())   # Output: 'hello'

# 8. Values of the following expressions
print('Remember, remember, the fifth of July.'.split())  
# Output: ['Remember,', 'remember,', 'the', 'fifth', 'of', 'July.']

print('-'.join('There can only one.'.split()))
# Output: 'There-can-only-one.'

# 9. Methods for right-justifying, left-justifying, and centering a string
text = "Hello"
print(text.rjust(10))  # Right-justify (adds spaces on the left)
print(text.ljust(10))  # Left-justify (adds spaces on the right)
print(text.center(10)) # Center (adds spaces on both sides)

# 10. Best way to remove whitespace characters from the start or end
whitespace_string = "   Hello, World!   "
print(whitespace_string.strip())  # Removes leading and trailing spaces
print(whitespace_string.lstrip()) # Removes leading spaces
print(whitespace_string.rstrip()) # Removes trailing spaces
```
